# Importing Datasets #

```python
import numpy as np
import pandas as pd
import re
dataSet  = pd.read_excel('Data/Data analyst Data.xlsx')
dataSet
```

# How many unique students are included in the dataset? #

```python
temp = dataSet['Designation'].value_counts().reset_index(name= 'Total Count')
temp.to_csv('Data/Basic Questions/Unique Intern.csv', index=False)
print(temp.to_string(index=False))
print('\n')
print("Total Number of Unique Interns Participated In This Program are: ", temp.shape[0])
```

# What is the average GPA of the students? #

```python
temp = dataSet['CGPA'].mean()
print("The Average CGPA of ALL Students is ",temp)
```

# What is the distribution of students across different graduation years? #

```python
temp = dataSet['Year of Graduation'].value_counts().reset_index(name='Total Count')
temp.to_csv('Data/Distibution on Year of Graduation.csv', index=False)
print(temp.to_string(index=False))
```

# What is the distribution of student's experience with Python programming? #

```python
temp = dataSet['Experience with python (Months)'].value_counts().reset_index(name='Total Count').sort_values(by='Experience with python (Months)').copy()
temp.to_csv('Data/Basic Questions/Experience with python.csv', index = False)
print(temp.to_string(index=False))
```

# What is the average family income of the student? #

```python
def extract_and_mean(income_string):
    matches = re.findall(r'\d+', income_string)
    numbers = [int(match) for match in matches]
    mean_value = sum(numbers) / len(numbers) if numbers else None
    return mean_value

def avgIncome(temp : pd.DataFrame) -> pd.DataFrame:
    a = temp.copy()
    x = temp['Family Income'].value_counts().reset_index(name='Total Count').sort_values(by='Family Income')
    x.to_csv('Data/Basic Questions/Average Income.csv', index=False)
    a['Assumed Income'] = a['Family Income'].apply(extract_and_mean)
    a['Assumed Income'].value_counts().reset_index(name='Count')
    print(x)
    print(f"The average income of Family of inturns is {a['Assumed Income'].mean()} Lakh") 

avgIncome(dataSet)
```

# How does the GPA vary among different colleges? (Show top 5 results only) #

```python
dataSet['CGPA'] = pd.to_numeric(dataSet['CGPA'], errors='coerce')
result = dataSet.groupby('College Name')['CGPA'].mean().reset_index()
result.to_csv('Data/Basic Questions/CGPA Vary among College.csv', index=False)
result.head(5)
```

# What is the average GPA for student from each city? #

```python
dataSet['CGPA'] = pd.to_numeric(dataSet['CGPA'], errors='coerce')
result = dataSet.groupby('City')['CGPA'].mean().reset_index()
result.to_csv('Data/Basic Questions/Average CGPA based on Cities.csv', index = False)
result
```

# Are there any outliers in the quantity (number of courses completed) attribute? #

```python
Q1 = dataSet['Quantity'].quantile(0.25)
Q3 = dataSet['Quantity'].quantile(0.75)
IQR = Q3 - Q1

outliers = dataSet[(dataSet['Quantity'] < Q1 - 1.5 * IQR) | (dataSet['Quantity'] > Q3 + 1.5 * IQR)]

if outliers.empty:
    print("No outliers found.")
else:
    print(f"{outliers} Outliers are found")
```
